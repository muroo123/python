### 01.range를 이용한 for문
- range() 함수를 사용하여 일련의 숫자 범위 생성

- stop, start, step은 정수
- range(stop) : 0 ~ stop -1까지의 범위
- range(start, stop) : start ~ stop-1 까지의 범위
- range(start, stop, step) : start ~ stop-1까지 step 씩 이동하는 범위
range는 수정이 불가능함


```python
list(range(10))
```

```python
# [1] 0 ~ 9를 한 줄에 출력하는 for문을 작성하라
for x in range(10):
    print(x, end=' ')
```

```python
# [2] 1~10까지 합계를 출력하는 for문을 작성하라
sum = 0
for x in range(1,10+1):
    sum = sum + x
    print(f'x값이 {x}일때 합계는 {sum}')
print(f'1부터 10까지 합계는 {sum}입니다.')
```

```python
# [3]입력으로 구구단 단을 입력받아 출력하기
# 2 -> 2단 출력하기

dan = int(input("몇단 출력할까요?"))
for x in range(1,10):
    print(f'{dan} * {x} = {dan*x}')
```

```python
# [3]숫자를 입력받아 해당 숫자까지 3의 배수 갯수 구하기

num3 = 0
num = int(input("숫자를 입력하세요?"))

for x in range(1, num+1):
    if x%3 == 0:
        num3 += 1
print(f'1부터 {num}까지 3의 배수는 {num3}개입니다.')
```

```python
# [4] 중첩 반복문 구구단 2~9단 출력
# 2~9단 for
# 1~9 for

for dan in range(2, 9+1):
    print(f'===={dan}단====')
    for x in range(1,10):
        print(f'{dan} * {x} = {dan*x}')
    print()
```

02. break, continue
- break : 반복동작을 중지
- continue : 반복을 계속하지만, 반복 내용 중 continue 다음의 문장을 처리하지 않음
  (다음 item을 가져오는 동작으로 바로 이동)


```python
# [1] 1~100합계를 구하다 합계가 1000이 넘으면 그만!!!  1000 초과 된 i값 찾기
sum = 0

for x in range(1,100+1):
    sum += x
    if sum > 1000 :
        break
print(f'1000초과 한 x값은 {x}이고 합계는 {sum}입니다.')
```

```python
# [2] 입력값까지 3의 배수를 제외한 합계 구하기
sum = 0
num = int(input("어디까지 더할까요?"))

for x in range(1,num+1):
    if x%3 == 0 :
        continue
    sum += x
print(f'1부터 {num}까지 3의 배수를 제외한 합계는 {sum}입니다.')
```

### 5-02.for문 연습문제 1(합격, 불합격)
- 총 5명 학생의 성적이 다음과 같을 때 시험 점수가 60점 이상이면 "합격입니다."를 그렇지 않으면 "불합격입니다."를 점수와 함께 출력하세요.
- 사용) result = [90, 35, 68, 44, 81]

```python
result = [90, 35, 68, 44, 81]

for score in result:
    if score >= 60 :
        print(f'{score} 합격입니다.')
    else :
        print(f'{score} 불합격입니다.')
```

```python
result = [90, 35, 68, 44, 81]

for score in result :
    msg='합격' if score >= 60 else '불합격'
    print(f'{score} {msg}입니다.')
```

### 5-03.for 고급

```python
# [1] for문을 사용하여 다음 a의 내용을 아래와 같이 출력하라
# 1 2
# 3 4
# 5 6
a = [(1, 2), (3, 4), (5, 6)]

for t,p in a:
    print(t,p)
```

```python
# [2] for문과 zip을 사용하여 아래와 같이 출력하라
# A 100
# B 200
# C 300
name = ['A', 'B', 'C']
unit = [100, 200, 300]

for n, u in zip(name, unit):
    print(n, u)
```

```python
# [3] for문과 enumerate을 사용하여 data의 값을 * 2 한 값으로 변경하여 출력하라
#     [20, 40, 60] 으로 수정하여 출력하도록 한다.
data = [10, 20, 30]
# (0, 10), (1, 20), (3, 30)

for i, d in enumerate(data):
    data[i] = d*2
print(data)

```

```python
data = [10, 20, 30]

for x in range(len(data)):
    data[x] = data[x] *2
print(data)
```

## 5-2. while statement

- 조건이 참인동안 계속 반복 수행하는 것
   - 조건이 거짓이 되는 것이 보장되어야 함
   - statement에서 break를 사용하여 종료할 수 있음


    while 조건식 :
        statement


image.png     

```python
# 1~100까지 합계구하기

sum = 0

for x in range(1,100+1) :
    sum += x
print(sum)
```

```python
sum = 0
x = 1

while x <= 100 :
    sum += x
    x += 1
print(sum)

```

```python
# 1~100 합계가 1000 초과이면 반복 탈출

sum = 0

for x in range(1,100+1):
    sum += x
    if sum > 1000:
        break
print(sum, x)
```

```python
sum = 0
x = 1

while x <= 100:
    sum += x
    if sum > 1000:
        break
    x +=1
print(sum,x)
```

```python
sum  = 0
x = 1
while True:
    sum += x
    if sum > 1000:
        break
    x +=1
print(sum,x)
```

```python
# [1] 암호를 입력 받아 틀린 암호인 경우 '암호가 잘못 입력 되었습니다.'를 출력하고 다시 입력 받는 것을 반복한다
#     올바른 암호가 입력될 때까지 계속 반복하도록 함
PW = 'bear'

while input("암호를 입력하세요 ") != PW :
    print("암호가 잘못 입력 되었습니다.")
print("문이 열렸습니다")
```

```python
# [2] 암호를 입력 받아 틀린 암호인 경우 '암호가 잘못 입력 되었습니다.'를 출력하고 다시 입력 받는 것을 반복한다
#     올바른 암호가 입력될 때까지 계속 반복하도록 함
PW = 'bear'

while True :
    if input("암호를 입력하세요") == PW :
        break
    print("암호가 잘못 입력 되었습니다.")

```
