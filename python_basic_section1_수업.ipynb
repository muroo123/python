### ◆ 표현식 (expression) 과 문장 (Statement)

### ◆ 명명규칙 (Naming Rule)
• 프로그래밍 에서 이름을 붙일 때 사용

영문, 숫자, _ (under score), 한글(비권장)

• 규칙
1. 키워드와 같은 이름을 사용할 수 없음(if, for, import)
2. 특수문자는 "_"만 사용 가능(#,@ 등 사용 안됨)
3. 숫자로 시작 할 수 없음(1a, 1b)
4. 문자 중간에 공백으로 시작할 수 없음(a b, ab c 등)
5. 표현하려는 것과 최대한 관련 있게(n -> cnt, a ->amt)
6. 다른 사람이 읽어도 바로 파악될 수 있게 해야 함.

## 1-03.서식있는 문자열
image.png

```python
# [1] 1000 값을 갖는 정수 객체 a를 사용하여 'Price is 1000'을 출력한다
# % 연산자, format 메서드, f-string 사용 => 문자열 만드는 방법
# %d : 10진 정수,  %f : 실수(float), %s : 문자열(str)

a = 1000
print('Price is', a)
print('Price is %d' % (a))
```

```python
# [2] 10, 20 값는 갖는 정수 객체 a, b를 사용하여
# '10 + 20 = 30'을 출력한다
# % 연산자, format 메서드, f-string 사용
a, b = 10, 20
print(a,'+', b, '=', a+b)
print('%d + %d = %d' % (a, b, a+b))
```

## 1-04.format() 함수 사용
•Python은 숫자를 문자열의 형태로 변환하여 출력할 수 있도록 format( ) 함수 제공

•새롭게 삽입될 자리에 중괄호{ }를 표시하고 출력될 값을 format(값)으로 적음.

image.png

```python
# [1] 1000 값을 갖는 정수 객체 a를 사용하여 'Price is 1000'을 출력한다
# % 연산자, format 메서드, f-string 사용 => 문자열 만드는 방법
# %d : 10진 정수,  %f : 실수(float), %s : 문자열(str)

a = 1000
print('Price is {}'.format(a))
print(f'Price is {a}')
```

```python
# [2] 10, 20 값는 갖는 정수 객체 a, b를 사용하여
# '10 + 20 = 30'을 출력한다
# % 연산자, format 메서드, f-string 사용
a, b = 10, 20
print('{} + {} = {}'.format(a, b, a+b))
print(f'{a} + {b} = {a+b}')
```
