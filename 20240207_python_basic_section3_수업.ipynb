## 3-4.연습문제2 - 학점구하기

````python
# 0~100점의 점수에 대해 다음 기준을 적용하여 학점을 출력한다
# 100 ~ 90 : A
# 89 ~ 80 : B
# 79 ~ 70 : C
# 69 ~ 60 : D
# 59 ~ 0 : F
# 단, 0점 미만, 100점 초과 점수에 대해서는 “Error” 를 출력한다
````

````python
score = 78

if score > 100 or score < 0 :
    msg = "error"
elif score >= 90 :
    msg = "A"
elif score >= 80:
    msg = "B"
elif score >= 70:
    msg ="C"
elif score >= 60:
    msg ="D"
else :
    msg ="F"
print(msg)
````

````python
# if ~ elif ~ elif ~ else 사용
score = -50

if score <= 100 and score >= 90 :
    msg = "A"
elif score <= 89 and score >= 80 :
    msg = "B"
elif score <= 79 and score >= 70 :
    msg = "C"
elif score <= 69 and score >= 60 :
    msg = "D"
elif score <= 59 and score >= 0 :
    msg = "F"
elif score > 100 or score < 0 :
    msg = "Error"

print(msg)
````

## 3-4.연습문제3 - 윤년 계산하기

````python
# 윤년 계산하기
# 윤년이란 태양년에 맞추기 위해서 4년에 한 번씩 윤날(2월 29일)을 추가하는 해를 말합니다.
# 4의 배수에 해당하는 해는 윤년이지만 100의 배수에 해당하는 해는 윤년에서 제외, 400배의 배수의 해는 윤년에 포함

# 입력 : 한 개의 정수가 입력된다.(조건문, 논리연산자,나머지 연산자를 사용하여 계산)
# 연도를 입력하세요 : 2019
# 출력 : 입력 받은 수가 윤년인지 아닌지 출력
# 2019 년은 윤년이 아닙니다.

# 연도 입력 받기
year = int(input("연도를 입력하세요 :"))

# 윤년 판단(4의 배수이고 100의 배수 아니다. 또는 400의 배수이다.)
if (year%4 == 0 and year%100 != 0) or (year%400 == 0) :
    print(f'{year}년은 윤년입니다.')
else :
    print(f'{year}년은 윤년이 아닙니다.')
````

주원이는 월, 수, 금, 일 아르바이트를 간다.

3 수요일

요일의 번호가 입력으로 주어지면 그 날이 아르바이트 가는 날이면 "oh my god"를 가는 날이 아니면 "enjoy"를 출력하시오.

````python
day = input("요일을 입력하세요")

if day > "7" or day < "1" :
    print("입력 오류!")
elif day == "1" or day == "3" or day == "5" or day == "7" :
    print("oh my god")
else :
    print("enjoy")
````
