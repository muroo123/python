## Section 7 : 예외처리

- 예외(exception)란 코드를 실행하는 중에 발생한 에러를 뜻합니다.
- 프로그램 실행 전(Syntax Error)
- 프로그램 실행 중(Runtime Error)


```python
# Syntax Error
print("Hello, World!)
```

```python
# Runtime Error : TypeError
print("Hello, world!" + 2020)
```


- try-exception : 오류가 발생할 경우 예외처리 구문 (try,except) 을 통해 예외 처리 할 수 있습니다.

    try:
        실행할 코드
    except:
        예외가 발생했을 때 처리하는 코드


```python
# [1] 숫자를 입력 받아 제곱을 계산해서 출력
# 숫자가 입력 되지 않으면 ValueError 발생

number = int(input("숫자를 입력하세요"))
print(number*number)
```

```python
# [2] try, except를 사용하여 숫자가 입력되지 않았을때 발생하는 error 처리
try:
    number = int(input("숫자를 입력하세요"))
    print(number*number)
except:    # 예외가 발생했을 때 실행됨
    print("숫자를 입력하지 않았습니다.")
```

- try-exception-else : 오류가 발생할 경우 예외처리 구문 (try,except)과 오류가 발생하지 않으면 실행 될 구분(else)을 구분해줌

    try:
        예외 가능성 코드
    except:
        예외가 발생했을 때 처리하는 코드
    else:
        예외가 없을 경우 실행하는 코드


```python
# [3] try, except,else 를 사용하여 처리 결과를 예외, 정상 구분하여 출력
try:
    number = int(input("숫자를 입력하세요"))
except:    # 예외가 발생했을 때 실행됨
    print("숫자를 입력하지 않았습니다.")
else:      # 예외가 없을 경우 실행 됨
    print(number*number)
```

- try-exception-else-finally
- try 블록 안에서의 오류 발생 여부와 상관 없이 무조건 실행
- 오류가 발생한 경우 해당 오류 이후는 실행이 되지 않음

    try:
        예외 가능성 코드
    except:
        예외가 발생했을 때 처리하는 코드
    else:
        예외가 없을 경우 실행하는 코드
    finally
        마지막에 실행 되는 코드

```python
# [4] try, except,else, finally
try:
    number = int(input("숫자를 입력하세요"))
except:    # 예외가 발생했을 때 실행됨
    print("숫자를 입력하지 않았습니다.")
else:      # 예외가 없을 경우 실행 됨
    print(number*number)
finally:   #마지막에 실행
    print("프로그램을 종료 합니다.")
```

- 예외 객체를 사용하여 Error 확인

    try:
        예외 가능성 코드
    except [예외종류] as [예외객체]:
        예외 경우 실행 코드

```python
# [5] 예외 객체 : 숫자를 입력 받아 제곱을 계산해서 출력
try:
    number = int(input ("숫자를 입력하세요") )
    print(number*number)
except Exception as exception:
    print("exception type: ", type(exception))
    print("exception: ", exception)
```

- 여러 개의 오류 처리하기

오류 종류에 따라 다른 except 문을 기술 할 수 있음.

except 문 기술 순서에 따라 최초 해당 하는 오류에 대해서만 처리됨

except (ValueError, IndexError, ZeroDivisionError) as exc : 도 가능


```python
#[6] 여러가지 오류 처리
my_list = [10, 20, 30, 40]
try:
    index = int(input("숫자를 입 력하세요 (0~3)"))
    value = my_list[index]
    print(index, my_list[index])
    #value.hi() #int type에서 지원 되지 않은 메소드 호출하면

except ValueError as exception:
    print("exception: ", exception)
    print("숫자를 입력해 주세요.")
except IndexError as exception:
    print("exception: ", exception)
    print("0~3 사이 숫자를 입력해 주세요.")
except Exception as exception:
    print("exception: ", exception)
    print("예상하지 못한 error")
```
