```python
# [4] random 난수 생성 모듈

import random

# 0.0 <= x <1.0 실수
print("random()  :", random.random())

#지정 범위 실수
print("uniform(5, 10)  : ", random.uniform(5, 10))

#0부터 지정 범위 정수
print("randrange(10)  : ", random.randrange(10))

#지정 범위 정수
print("randrange(5, 10)  : ", random.randrange(5, 10))

#지정 범위 정수 5 <= x <= 10
print("randint(5, 10)  : ", random.randint(5, 10))

#지정 값 중 하나 선택
print("choice([1,2,3,4,5])", random.choice([1,2,3,4,5]))

#지정 값에서 k개 임의의 원소 추출(unique)
print("sample([1,2,3,4,5])", random.sample([1,2,3,4,5], 3))
```

```python
# [6]datatime 모듈 : date(날짜), time(시간) 관련 모듈

import datetime

now = datetime.datetime.now()
print(now)
print (now.year, '년' )
print (now.month, '월' )
print (now.day, '일' )
print (now.hour, '시')
print (now.minute, '분')
print (now.second, '초')
```

```python
# [7]time 모듈 : 시간 관련된 모듈
import time

# time.time()은 UTC(Universal Time Coordinated 협정 세계 표준시)를 이용하여 현재 시간을 실수 형태로 리턴하는 함수.
# 1970년 1월 1일 0시 0분 0초를 기준으로 지난 시간을 초 단위로 리턴
print(time.time())

#time.localtime은 time.time()에 의해서 반환된 실수값을 이용해서 연도, 월, 일, 시, 분, 초... 형태로 변환해 주는 함수
print(time.localtime(time.time()))

#time.localtime으로 변환 된 값을 보기 좋게 출력해주는 함수
print(time.asctime(time.localtime(time.time())))
print(time.ctime())

#time.sleep 일정시간 간격을 두고 실행
# 1초 간격으로 반복문 실행
for i in range(10):
    print(i)
    time.sleep(1)


```

```python
# [8]프로그램 실행에 걸리 시간 체크
import time
before = time.time()
my_list = range(10000000)
cp_list = []

for i in my_list:
    cp_list.append(i)
after = time.time()
diff = after - before
print ("처리시간: %.2f s" % diff)
```

### [연습문제] 로또(Lotto) 번호 생성기

```python
# [1] 1~45 범위 choice를 사용하여 하나씩 선택
import random
lotto_no=[x for x in range(1,46)]
lotto =[]

for _ in range(6):
    choice = random.choice(lotto_no)
    lotto.append(choice)
    lotto_no.remove(choice)

print(lotto)
print(lotto_no)
```

```python
# [2] 1~45 범위의 숫자에서 6개 데이터 Sampling
import random
#random.seed(1)
lotto = random.sample(range(1, 46), 6)
print(lotto)
```
