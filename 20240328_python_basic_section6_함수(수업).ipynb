### 6-04. argument의 종류
- 함수의 호출시 사용되는 argument    
    - positional argument : 위치에 따라 값이 전달되는 방식
    - keyword argument : = 를 사용하여 특정 parameter에 값을 제공함, 순서에 상관 없음
- 함수 정의시 사용되는 argument
    - default argument : 새로운 argument가 전달되지 않을 경우 사용되는 기본 값
- Python의 argument rule
     1. 함수 정의 시 : non-default argument가 앞, default argument가 뒤에 위치해야 한다.
     2. 함수 호출 시 : positional argument가 앞, keyword argument가 뒤에 위치해야 한다.

````python
def function(a, b):  # a, b : parameter
    return [a, b]

# 어떻게 사용하는 것일까요?
x = function(10, 20)       # positional argument - 순서에 의해서 parameter에 argument를 전달하는 방식
y = function(b=10, a=20)   # keyword argument - parameter를 지정하여 argument를 전달하는 방식
print(x, y)
````

````python
# keyword only parameter
def func(*, a, b) :
    print(a)

func(b=20, a=10)    # keyword argument
# func(10, 20)  # Error
````

````python
# a : ( positional or keyword) parameter
# b, c : ( keyword only ) parameter
def func(a, *, b, c) :
  return [a, b, c]   # ( a = 10, b = 20, c = 30)

r1 = func(10, b=20, c=30)    # a를 positional argument
r2 = func(a=10, b=20, c=30)  # a를 keyword argument
r3 = func(c=30, b=20, a=10)   # keyword argument 일때는 순서가 중요하지 않음
print(r1, r2, r3)
````

가변 파라미터

**var-positional parameter**  ```(*a)```
- parameter 이름에 *가 하나 붙어 있음
- tuple 형태로 전달된다

**var-keyword parameter**  ```(**a)```
- parameter 이름에 *이 두개 붙어 있음
- dict 형태로 전달된다

**var positional, var keyword**
- var는 0개 이상의 가변적 argument를 의미한다
- default argument를 지정할 수 없음
- var keyword parameter는 단독으로 사용되거나, 맨 마지막 parameter로만 사용

````python
# var positional parameter
# a의 type은 무엇일까요? tuple

def func (*a):
    print(type(a), a)

func(1,2,3,4,5,6,7)
func()
````

````python
# 함수, 이름 몇번 출력하기
def hello(cnt, *name):
    for n in name:
        print(n * cnt)

hello(2,"Alice","Rabbit","Tom")
````

````python
# a : (var - positional) parameter - default argument를 갖지 못함
# b : ( keyword only ) parameter
def func (*a, b):
    print(type(a), a)
    print(b)

func(1,2,3,4,5,6,7, b=10)
func(b=20)
````

````python
# var-keyword parameter : 맨 마지막 parameter로만 사용가능
# a의 type은 무엇일까요? dict
def func(**a):
    print(type(a), a)

func(A=10, B=20, C=30, D=40)
func()
````

### 함수 연습

````python
# [1] square 함수 작성
# • 1개매개변수
# • 매개변수를 제곱하여 그 값을 return
# • square(3) 으로 호출 했을 때 9가 출력

def square(n):
    return n*n

value = square(3)
print(value)
````

````python
# [2] 3과목의 평균을 계산하는 avg
# • 파라미터 3개
# • 3개의 평균계산하여 리턴
# • average(70, 30, 50)

def average(a,b,c):
    total = a+b+c
    return int(total/3)

avg = average(70,30,50)
print(avg)
````

````python
# [3] N개 과목의 평균을 계산하는 avg
# • 파라미터수 제한없음
# • 모든 과목의 점수에 대한 평균 점수를 리턴
# • average(70, 30, 50, 60, 70, 80)

def average(*score):
    total = sum(score)
    avg = total/len(score)
    return avg

avg = average(70, 30, 50, 60, 70, 80)
print(avg)
````

````python
# [4] N개 과목의 평균을 grade로 출력
# • A (90점 이상), B (80점 이상), C (70점 이상), 나머지 F
def get_grade(avg):
    if avg >= 90:
        return 'A'
    elif avg >= 80:
        return 'B'
    elif avg >= 70 :
        return 'C'
    else :
        return 'F'

avg = average(70, 30, 50, 60, 70, 80)
grade = get_grade(avg)
print(avg)
print(grade)
````

````python
# [3] 두 list의 합을 구하는 식 작성 (map, lambda 사용)
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]
# [3, 6, 9, 12, 15]
c = list(map(lambda x, y: x+y, a, b))
print(c)
````

````python
# [4] 두 list의 합을 구하는 식 작성 (zip, comprehension 사용)
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]
c = [x+y for x, y in zip(a, b)]
print(c)
````

````python
# [5] age 리스트의 모든 숫자에 대해 1의 자리를 0으로 변경
age = [23, 45, 38, 51]
r = list(map(lambda x: x//10*10, age))
print(r)
````
